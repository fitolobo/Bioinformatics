# Ejemplo 2: Cargando un objeto fasta

La documentación de la librería y otras librerías que hemos encontrado hasta el momento en ```python``` las puedes encontrar en los siguientes links:

5 - [Ejemplo de un Fasta File](https://www.ncbi.nlm.nih.gov/nuccore/AY804513.1?report=fasta)

Tranqui!, la próxima clase vamos a descargar todo esto en tu computador y cambiar las rutas para que funcione.

````python
import fastaparser #importamos la librería
````

````python
with open("./datos/sample.fasta") as fasta_file: #se utiliza un manejador de contexto
        parser = fastaparser.Reader(fasta_file)      #la librería "parsea" o lee el fasta file
        for seq in parser:                           #como trae varios objetos llamamos uno a uno lo que hay adentro
            # Imprimimos lo que habia adentro de cada parte del fastafile
            print('ID:', seq.id)
            print('Description:', seq.description)
            print('Sequence:', seq.sequence_as_string())
            print()
````
